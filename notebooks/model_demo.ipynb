# Model Demonstration

This notebook demonstrates the end-to-end pipeline of the **Stock Sentiment Predictor**.

```python
import sys
import os
sys.path.append(os.path.abspath('../src'))

import pandas as pd
import matplotlib.pyplot as plt
from data_pipeline import fetch_stock_data, fetch_stock_news, simulate_historical_news
from sentiment_analysis import load_sentiment_model, process_news_sentiment, aggregate_daily_sentiment
from model_training import prepare_data, train_models
from evaluation import evaluate_models
```

## 1. Data Collection

```python
ticker = 'AAPL'
start_date = '2023-01-01'
end_date = '2023-12-31'

# Fetch historical price data
df_prices = fetch_stock_data(ticker, start_date=start_date, end_date=end_date)

# For demonstration purposes, we will simulate historical news as most APIs for historical news are paid.
simulate_historical_news(ticker, start_date=start_date, end_date=end_date)
```

## 2. Sentiment Analysis

```python
# Load FinBERT
sentiment_pipe = load_sentiment_model()

# Process simulated news
target_file = f"../data/raw_data/{ticker}_simulated_news.json"
sentiment_df = process_news_sentiment(ticker, target_file, sentiment_pipe)

df_daily_sentiment = aggregate_daily_sentiment(sentiment_df, ticker)
df_daily_sentiment.head()
```

## 3. Model Training & Evaluation

```python
# Prepare data by merging and engineering features
df_merged = prepare_data(ticker)

# Train Random Forest & Logistic Regression
df_test_results = train_models(df_merged, ticker)

# Print metrics and generate charts
evaluate_models(ticker)
```
