# Data Exploration Studio

This notebook provides an exploratory data analysis (EDA) environment for the stock prices and simulated sentiment scores.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="whitegrid")
```

## Load Sample Data
Make sure you have run the data pipeline (`src/data_pipeline.py`) before exploring the data.

```python
ticker = "AAPL"
price_file = f"../data/raw_data/{ticker}_prices.csv"
sentiment_file = f"../data/processed_data/{ticker}_daily_sentiment.csv"

try:
    df_price = pd.read_csv(price_file)
    df_price['Date'] = pd.to_datetime(df_price['Date'])
    print("Price data loaded successfully.")
    display(df_price.head())
except FileNotFoundError:
    print(f"Could not find {price_file}. Please run the data pipeline.")
```

```python
try:
    df_sentiment = pd.read_csv(sentiment_file)
    df_sentiment['date'] = pd.to_datetime(df_sentiment['date'])
    print("Sentiment data loaded successfully.")
    display(df_sentiment.head())
except FileNotFoundError:
    print(f"Could not find {sentiment_file}. Run the sentiment analysis script.")
```

## Visualizing Historical Prices

```python
if 'df_price' in locals():
    plt.figure(figsize=(12, 6))
    plt.plot(df_price['Date'], df_price['Close'], label='Close Price', color='b')
    plt.title(f"{ticker} Historical Close Price")
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.show()
```

## Sentiment Distribution

```python
if 'df_sentiment' in locals():
    plt.figure(figsize=(8, 5))
    sns.histplot(df_sentiment['sentiment_score'], bins=20, kde=True, color='purple')
    plt.title('Distribution of Daily Sentiment Scores')
    plt.xlabel('Sentiment Score (-1 to 1)')
    plt.ylabel('Frequency')
    plt.show()
```
